```python
import numpy as np

# Load the .npy file
training_loss = np.load('../../output/alpaca-gpt4_20000_fedavg_c20s2_i10_b16a1_l512_r32a64_20250523232123/training_loss.npy')
```

```python
# Check the shape
print("Shape of training_loss:", training_loss.shape)  # Should be (2, 200)

# Plotting
plt.figure(figsize=(12, 6))
num_clients = training_loss.shape[0]
num_rounds = training_loss.shape[1]

for client_id in range(num_clients):
    client_loss = training_loss[client_id]
    
    # Mask the -1 entries (non-participation) for plotting
    rounds = np.arange(num_rounds)
    mask = client_loss != -1
    plt.plot(rounds[mask], client_loss[mask], label=f"Client {client_id}")

plt.title("Training Loss per Round for Each Client")
plt.xlabel("Federated Round")
plt.ylabel("Training Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Mask out the -1 values (non-participation)
valid_losses = training_loss[training_loss != -1]

# Compute the average
average_loss = np.mean(valid_losses)

print(f"Average Training Loss (across all clients and rounds): {average_loss:.4f}")
```

```python
# Convert -1 (non-participation) to np.nan to ignore in average
training_loss_with_nan = np.where(training_loss == -1, np.nan, training_loss)

# Compute average loss per round (ignoring NaNs)
avg_loss_per_round = np.nanmean(training_loss_with_nan, axis=0)

# Plotting
plt.figure(figsize=(12, 6))

# Plot individual client losses
num_clients, num_rounds = training_loss.shape
rounds = np.arange(num_rounds)

for client_id in range(num_clients):
    client_loss = training_loss_with_nan[client_id]
    plt.plot(rounds, client_loss, label=f"Client {client_id}", alpha=0.6)

# Plot average loss per round
plt.plot(rounds, avg_loss_per_round, label="Average Loss", color="black", linewidth=2.5)

# Formatting
plt.title("Client-wise and Average Training Loss per Round for FedAvg without Differential Privacy")
plt.xlabel("Federated Round")
plt.ylabel("Training Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import os

# === CONFIGURE BASE PATH ===
base_path = "../../../OpenFedLLM/server_trained_files/"  # or output/

# === Track minimum ===
min_loss_overall = float('inf')
min_info = ""

# === Search recursively ===
for root, dirs, files in os.walk(base_path):
    if "training_loss.npy" in files:
        loss_path = os.path.join(root, "training_loss.npy")
        try:
            training_loss = np.load(loss_path)
            training_loss = np.where(training_loss == -1, np.nan, training_loss)
            avg_loss_per_round = np.nanmean(training_loss, axis=0)

            round_min = np.nanmin(avg_loss_per_round)
            if round_min < min_loss_overall:
                min_loss_overall = round_min
                min_info = f"{loss_path} (Round {np.nanargmin(avg_loss_per_round)})"

        except Exception as e:
            print(f" Failed to process {loss_path}: {e}")

# === Result ===
if min_info:
    print(f" Lowest Avg Training Loss: {min_loss_overall:.4f}")
    print(f" Found in: {min_info}")
else:
    print(" No valid training_loss.npy files found.")
```

```python
import numpy as np
import os

# === CONFIGURE BASE PATH ===
base_path = "../../../OpenFedLLM/output"
top_n = 5  # Change this to see more or fewer entries

# === Store all (loss, path, round) entries ===
all_losses = []

# === Search recursively for training_loss.npy ===
for root, dirs, files in os.walk(base_path):
    if "training_loss.npy" in files:
        loss_path = os.path.join(root, "training_loss.npy")
        try:
            training_loss = np.load(loss_path)
            training_loss = np.where(training_loss == -1, np.nan, training_loss)
            avg_loss_per_round = np.nanmean(training_loss, axis=0)

            round_min = np.nanmin(avg_loss_per_round)
            round_idx = np.nanargmin(avg_loss_per_round)

            all_losses.append((round_min, loss_path, round_idx))

        except Exception as e:
            print(f" Failed to process {loss_path}: {e}")

# === Sort and display top N ===
if all_losses:
    sorted_losses = sorted(all_losses, key=lambda x: x[0])
    print(f" Top {min(top_n, len(sorted_losses))} Lowest Avg Training Losses Found:")
    for i, (val, path, rnd) in enumerate(sorted_losses[:top_n], 1):
        print(f"{i}.  Loss: {val:.4f} | Round: {rnd} | Path: {path}")
else:
    print(" No valid training_loss.npy files found in the output/ directory.")
```
