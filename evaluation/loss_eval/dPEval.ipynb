```python
import numpy as np
import matplotlib.pyplot as plt

path = "../../../OpenFedLLM/server_trained_files/output_dp/fedadam_epsilon25.0_clip0.1/alpaca-gpt4_20000_fedadam_c20s2_i10_b16a1_l512_r32a64_20250527122359/training_loss.npy"
# Load training loss
training_loss = np.load(path)  # Update with your path

# Replace -1 with NaN to ignore non-participation
training_loss_with_nan = np.where(training_loss == -1, np.nan, training_loss)

# Compute average loss per round
avg_loss_per_round = np.nanmean(training_loss_with_nan, axis=0)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(avg_loss_per_round, color="blue", linewidth=2)
plt.title("Average Training Loss per Round (FedAdam, Differential Privacy(ε=25.0, Clip Norm=0.1))")
plt.xlabel("Epochs")
plt.ylabel("Average Training Loss")
plt.ylim(0.7, 1.35)  # Adjust upper limit as needed
plt.grid(True)
plt.tight_layout()
plt.show()
```
